```python
import numpy
precision = [0.9885103 , 0.99408721, 0.99716178, 0.97678221, 0.99534247,
       0.99420443, 0.98779637, 0.98921569, 0.99586777, 0.97588832,
       0.99332963]
sensitivity =[0.99720224, 0.9962963 , 0.99246704, 0.97806156, 0.97871767,
       0.98589342, 0.99683322, 0.9990099 , 0.9790625 , 0.98779705,
       0.99388209]
```

```python
mean1 = sum(precision)/len(precision)
mean2 = sum(sensitivity)/len(sensitivity)
print(mean1,mean2)
```

```python
import matplotlib.pyplot as plt
categories = []
for i in range(1,12):
    categories.append(i)

plt.figure(figsize=(4,3), dpi=300) 
plt.bar(categories,precision,color = 'gray')
plt.xlabel('Genus')
plt.ylabel('Precision')
plt.xticks(categories)
plt.ylim(0.6,1.01)
plt.gca().spines['right'].set_color('none')  
plt.gca().spines['top'].set_color('none') 
plt.savefig('precision.pdf', format='pdf')  
```
