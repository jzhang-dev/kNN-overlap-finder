```python
overlap_candidates = []
with open('/home/miaocj/lowhash.csv', 'r') as f:
    for lines in f:
        line = lines.strip().split('\t')
        r1_idx = name_id_dict[line[0]]
        if line[2] == '0':
            r2_idx = name_id_dict[line[1]]
        else:
            r2_idx = name_id_dict[line[1]]+1
        overlap_candidates.append((r1_idx, r2_idx))
```

```python
ref_graph_path = "/home/miaocj/docker_dir/kNN-overlap-finder/data/regional_reads/CHM13/all/filter3_real_new_wy/reference_graph.gpickle"
with open(ref_graph_path,'rb') as f:
    reference_graph = pickle.load(f)
```

```python
for overlap_candicate in overlap_candidates:
    if overlap_candicate in reference_graph.edges():
        print(overlap_candicate)
```
